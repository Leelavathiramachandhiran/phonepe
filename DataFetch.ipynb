```python
import streamlit as st
import os
import json
import pandas as pd
import pyodbc
import mysql.connector
```

```python
#aggregate_insurance
path_1 ="C:/Users/Documents/Project_2/pulse/data/aggregated/insurance/country/india/state/"
aggre_insurance_list = os.listdir(path_1)

column_1={"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}

for state in aggre_insurance_list:
    present_states=path_1+state+"/"
    aggre_year_list=os.listdir(present_states)
    
    for year in aggre_year_list:
        present_year=present_states+year+"/"
        aggre_file_list=os.listdir(present_year)
        
        for file in aggre_file_list:
            present_file=present_year+file
            data=open(present_file, "r")
            k=json.load(data)
            

            for i in k["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                column_1["Transaction_type"].append(name)
                column_1["Transaction_count"].append(count)
                column_1["Transaction_amount"].append(amount)
                column_1["States"].append(state)
                column_1["Years"].append(year)
                column_1["Quarter"].append(int(file.strip(".json")))
                

aggre_insurance=pd.DataFrame(column_1)


print(aggre_insurance)
```

```python
#AGGREGATE_TRANSACTION
path_2="C:/Users/Documents/Project_2/pulse/data/aggregated/transaction/country/india/state/"

aggre_trans_list = os.listdir(path_2)
 
column_2={"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}

for state in aggre_trans_list:
    present_states=path_2+state+"/"
    aggre_year_list=os.listdir(present_states)
    
    for year in aggre_year_list:
        present_year=present_states+year+"/"
        aggre_file_list=os.listdir(present_year)
        
        for file in aggre_file_list:
            present_file=present_year+file
            data=open(present_file, "r")
            S=json.load(data)

            for i in S["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                column_2["Transaction_type"].append(name)
                column_2["Transaction_count"].append(count)
                column_2["Transaction_amount"].append(amount)
                column_2["States"].append(state)
                column_2["Years"].append(year)
                column_2["Quarter"].append(int(file.strip(".json")))

aggre_transaction=pd.DataFrame(column_2)

aggre_transaction["States"] = aggre_transaction["States"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")

aggre_transaction["States"] = aggre_transaction["States"].str.replace("-"," ")
aggre_transaction["States"] = aggre_transaction["States"].str.title()
aggre_transaction["States"] = aggre_transaction["States"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","Dadra and Nagar Haveli and Daman Diu")
print(aggre_transaction)
```

```python
#Aggregate_User
path_3="C:/Users/Documents/Project_2/pulse/data/aggregated/user/country/india/state/"
aggre_user_list = os.listdir(path_3)

column_3 = {"States":[], "Years":[], "Quarter":[], "Brands":[], "Transaction_count":[], "Percentage":[]}

for state in aggre_user_list:
    present_states = path_3+state+"/"
    aggre_year_list = os.listdir(present_states)
    
    for year in aggre_year_list:
        present_year = present_states+year+"/"
        aggre_file_list = os.listdir(present_year)
        
        for file in aggre_file_list:
            present_file = present_year+file
            data = open(present_file, "r")
            U = json.load(data)
            

            try:

                for i in U["data"]["usersByDevice"]:
                    
                    brand = i["brand"]
                    count =i["count"]
                    percentage = i["percentage"]
                    column_3["Brands"].append(brand)
                    column_3["Transaction_count"].append(count)
                    column_3["Percentage"].append(percentage)
                    column_3["States"].append(state)
                    column_3["Years"].append(year)
                    column_3["Quarter"].append(int(file.strip(".json")))
            except:
                pass
            
aggre_user=pd.DataFrame(column_3)

aggre_user["States"] = aggre_user["States"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")

aggre_user["States"] = aggre_user["States"].str.replace("-"," ")
aggre_user["States"] = aggre_user["States"].str.title()
aggre_user["States"] = aggre_user["States"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","Dadra and Nagar Haveli and Daman Diu")


print(aggre_user)
```

```python
#Map_hover
#Map_insurance
path_4="C:/Users/Documents/Project_2/pulse/data/map/insurance/hover/country/india/state/"
map_hover_list = os.listdir(path_4)

column_4 = {"States":[], "Years":[], "Quarter":[],"Transaction_count":[], "Transaction_amount":[], "District":[]}

for state in map_hover_list:
    present_states = path_4+state+"/"
    map_year_list = os.listdir(present_states)
    
    for year in map_year_list:
        present_year = present_states+year+"/"
        map_file_list = os.listdir(present_year)
        
        for file in map_file_list:
            present_file = present_year+file
            data = open(present_file, "r")
            H= json.load(data)
            
            for i in H["data"]["hoverDataList"]:
                name = i["name"]
                count = i["metric"][0]["count"]
                amount = i["metric"][0]["amount"]
                column_4["District"].append(name)
                column_4["Transaction_count"].append(count)
                column_4["Transaction_amount"].append(amount)
                column_4["States"].append(state)
                column_4["Years"].append(year)
                column_4["Quarter"].append(int(file.strip(".json")))
map_hover=pd.DataFrame(column_4)

map_hover["States"] = map_hover["States"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")

map_hover["States"] = map_hover["States"].str.replace("-"," ")
map_hover["States"] = map_hover["States"].str.title()
map_hover["States"] = map_hover["States"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","Dadra and Nagar Haveli and Daman Diu")
print(map_hover)
```

```python
#Map_transaction 
path_5 = "C:/Users/Documents/Project_2/pulse/data/map/transaction/hover/country/india/state/"
map_trans_list = os.listdir(path_5)

column_5 = {
    "States": [],
    "Years": [],
    "Quarter": [],
    "District": [],
    "Transaction_count": [],
    "Transaction_amount": []
}

for state in map_trans_list:
    present_states = path_5 + state + "/"
    map_year_list = os.listdir(present_states)

    for year in map_year_list:
        present_year = present_states + year + "/"
        map_file_list = os.listdir(present_year)

        for file in map_file_list:
            present_file = present_year + file
            with open(present_file, "r") as data:
                S = json.load(data)

            
                for i in S["data"]["hoverDataList"]:
                    column_5["District"].append(i["name"])
                    column_5["Transaction_count"].append(i["metric"][0]["count"])
                    column_5["Transaction_amount"].append(i["metric"][0]["amount"])
                    column_5["States"].append(state)
                    column_5["Years"].append(year)
                    column_5["Quarter"].append(int(file.replace(".json", "")))

# Create DataFrame
map_transaction = pd.DataFrame(column_5)

map_transaction["States"] = map_transaction["States"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")

map_transaction["States"] = map_transaction["States"].str.replace("-"," ")
map_transaction["States"] = map_transaction["States"].str.title()
map_transaction["States"] = map_transaction["States"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","Dadra and Nagar Haveli and Daman Diu")
print(map_transaction)
```

```python
#Map_user
path_6="C:/Users/Documents/Project_2/pulse/data/map/user/hover/country/india/state/"
map_user_list = os.listdir(path_6)

column_6 = {"States":[], "Years":[], "Quarter":[], "Districts":[], "RegisteredUser":[], "AppOpens":[]}

for state in map_user_list:
    present_states = path_6+state+"/"
    map_year_list = os.listdir(present_states)
    
    for year in map_year_list:
        present_year = present_states+year+"/"
        map_file_list = os.listdir(present_year)
        
        for file in map_file_list:
            present_file=present_year+file
            data=open(present_file, "r")
            L = json.load(data)
            
            for i in L["data"]["hoverData"].items():
                district = i[0]
                registereduser = i[1]["registeredUsers"]
                appopens = i[1]["appOpens"]
                column_6["Districts"].append(district)
                column_6["RegisteredUser"].append(registereduser)
                column_6["AppOpens"].append(appopens)
                column_6["States"].append(state)
                column_6["Years"].append(year)
                column_6["Quarter"].append(int(file.strip(".json")))

map_user = pd.DataFrame(column_6)


map_user["States"] = map_user["States"].str.replace("andaman-&-nicobar-islands","Andaman and Nicobar")

map_user["States"] = map_user["States"].str.replace("-"," ")
map_user["States"] = map_user["States"].str.title()
map_user["States"] = map_user["States"].str.replace("dadra-&-nagar-haveli-&-daman-&-diu","Dadra and Nagar Haveli and Daman Diu")
print(map_user)
```
